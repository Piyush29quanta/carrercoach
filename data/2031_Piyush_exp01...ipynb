```python
import matplotlib.pyplot as plt
import numpy as np
```

Unit Step

```python
def unit_step(t):
  return np.heaviside(t,1)
```

```python
t = np.linspace(-5,5,100)
y = unit_step(t)
```

```python
plt.figure()
plt.plot(t,y)
plt.title('Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

Unit Ramp


```python
def unit_ramp(t):
  return np.where(t>=0,t,0)
```

```python
t = np.linspace(-5,5,100)
y = unit_ramp(t)
```

```python
plt.figure()
plt.plot(t,y)
plt.title('Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

Impulse

```python
def unit_impulse(t):
  return np.where(t==0,1,0)
```

```python
t = np.linspace(-5,5,100)
y = unit_impulse(t)
```

```python
plt.figure()
plt.plot(t,y)
plt.title('Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.figure()
plt.stem([0],[1], linefmt='--', markerfmt='ro', basefmt='blue')
plt.xlim(-5,5)
plt.ylim(-0.2,1.2)
plt.show()
```
